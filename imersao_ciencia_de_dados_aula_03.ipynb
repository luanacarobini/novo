```python
mapa = {'droga': 'composto'}
dados.rename(columns=mapa, inplace=True)
```

```python
dados.head
```

```python
cod_compostos = dados['composto'].value_counts().index[0:5]
```

```python
cod_compostos
```

```python
#Dentro da coluna 'composto', tudo o que estiver contido na variável cod_compostos, retorna o data frame apenas com as informações de cod_compostos.
#O @ é usado para puxar uma informação fora da string.
#Aqui estão sendo identificados as id's, ou seja, as culturas de células, em que estão os compostos mais usados.


dados.query('composto in @cod_compostos')
```

```python
#Gráfico feito com os 5 compostos mais usados.
#sns.set() - usado para estilizar o gráfico com configurações do seaborn.
#matplotlip.plot vai ajustar alguns padrões de tamanho, etc.
#plt.figure(figsize=(10,8)) - tamanho da figura.
#ax - variável para chamar o gráfico.
#plt.show() - tira informações externas ao gráfico.

import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
plt.figure(figsize=(8,6))
ax = sns.countplot(x='composto', data=dados.query('composto in @cod_compostos'))
ax.set_title('Top 5 Compostos')
plt.show()
```
