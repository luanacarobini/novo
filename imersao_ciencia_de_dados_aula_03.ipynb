```python
mapa = {'droga': 'composto'}
dados.rename(columns=mapa, inplace=True)
```

```python
dados.head
```

```python
cod_compostos = dados['composto'].value_counts().index[0:5]
```

```python
cod_compostos
```

```python
#Dentro da coluna 'composto', tudo o que estiver contido na variÃ¡vel cod_compostos, retorna o data frame apenas com as informaÃ§Ãµes de cod_compostos.
#O @ Ã© usado para puxar uma informaÃ§Ã£o fora da string.
#Aqui estÃ£o sendo identificados as id's, ou seja, as culturas de cÃ©lulas, em que estÃ£o os compostos mais usados.


dados.query('composto in @cod_compostos')
```

```python
#GrÃ¡fico feito com os 5 compostos mais usados.
#sns.set() - usado para estilizar o grÃ¡fico com configuraÃ§Ãµes do seaborn.
#matplotlip.plot vai ajustar alguns padrÃµes de tamanho, etc.
#plt.figure(figsize=(10,8)) - tamanho da figura.
#ax - variÃ¡vel para chamar o grÃ¡fico.
#plt.show() - tira informaÃ§Ãµes externas ao grÃ¡fico.

import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
plt.figure(figsize=(8,6))
ax = sns.countplot(x='composto', data=dados.query('composto in @cod_compostos'))
ax.set_title('Top 5 Compostos')
plt.show()
```
